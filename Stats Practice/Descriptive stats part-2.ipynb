```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bernoulli's distribution
from scipy.stats import bernoulli
bd = bernoulli(0.6) # p=0.6,the probability of success
x = [0,1] #outcome of experiment can take value as 0,1
```

```python
#lets create a bar plot
plt.figure(figsize=(12,8))
sns.barplot(x=x,y=bd.pmf(x))
```

```python
# Poisson distribution
plt.figure(figsize=(12,8))
sns.histplot(np.random.poisson(lam=5,size=100),kde=True)
```

```python
# Uniform distribution
from scipy.stats import uniform
x = np.linspace(uniform.ppf(0.01),
                uniform.ppf(0.99), 100)
rv = uniform()
plt.plot(x, rv.pdf(x), 'k-', lw=2)
```

```python
r = uniform.rvs(size=1000)
sns.histplot(data=r, alpha=0.8,kde=True)
```

```python
# 2 - Continuous probability distributions
```

```python
# Normal distribution / Gaussian distribution

# we already have a normally distributed data
df = sns.load_dataset('iris')
data = df['sepal_width']
sns.histplot(data,kde=True)
```

```python
# Continuous uniform distribution
sns.histplot(np.random.uniform(size=1000),kde=True,color='red',alpha=0.5)
```

```python
# Log normal distribution
# the random variables whose log values follow  a normal distribution
from scipy.stats import lognorm 
x = np.linspace(0, 5, 100)
     
# Varying positional arguments 
y1 = lognorm.pdf(x, 1, 3) 
y2 = lognorm.pdf(x, 1, 4) 
plt.plot(x, y1, "*", x, y2, "r--") 
```

```python
# Exponential distribution
plt.figure(figsize=(9,6))
ex_dist = np.random.exponential(3,1000)
sns.histplot(ex_dist,kde=True,color='green',alpha=0.4)
```

```python
# standard normal distribution
# gaussian to standard normal distribution
# standardization
# basically we use zscore formula for every data element
new_df = df.copy()
new_df['standard_normal_data'] = [(i-new_df['sepal_width'].mean())/new_df['sepal_width'].std() for i in new_df['sepal_width']]
new_df
```

```python
sns.histplot(new_df['standard_normal_data'],kde=True) # the standard normal data have mean = 0 and standard deviation of 1
```

```python
# Normalization
# we have used minmaxscaler function for normalization
new_df['normalization_data'] = [(i-min(new_df['sepal_width']))/(max(new_df['sepal_width'])-min(new_df['sepal_width'])) for i in new_df['sepal_width']]
```

```python
sns.histplot(new_df['normalization_data'],kde=True) # normalized data always lies between 0-1
```
